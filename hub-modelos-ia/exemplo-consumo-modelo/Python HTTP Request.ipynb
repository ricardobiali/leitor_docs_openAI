## Exemplo de uso Python via requisição HTTP

````python
import requests

url = 'https://apid.petrobras.com.br/ia/openai/v1/openai-azure/openai/deployments/gpt-4-petrobras/chat/completions?api-version=2023-12-01-preview'
headers = {
    'Content-Type': 'application/json',
    'api-key': '<api-key para a sua chave pessoal ou de sistema>'
}
data = {
	"messages": [
		{
			"role": "system",
			"content": "You are a helpful assistant."
		},
		{
			"role": "user",
			"content": "Como faço para chamar uma API REST usando Python?"
		}
	]
}

response = requests.post(url, headers=headers, json=data)

print(response.status_code)
print(response.json())
````
